```python
pop_trends = pd.read_csv('population trends.csv', thousands=',')
pop_trends

```

```python
pop_trends = pop_trends.rename(columns={"Country": "country" , "2019": "Population in Millions in 2019", "2030": "Projected Population in Millions in 2030"})
```

```python
pop_trends.dtypes
```

```python
# pd.Series(['260,327,021']).str.replace(',', '').astype(int)

# Retypes columns for joining.
pop_trends['Population in Millions in 2019'] = pop_trends['Population in Millions in 2019'].astype(float).astype(int)
pop_trends['Projected Population in Millions in 2030'] = pop_trends['Projected Population in Millions in 2030'].astype(float).astype(int)

```

```python
# # result = pd.merge(left, right, on="key")
# final_dataset = pd.merge(nfa, pop_trends, on="country", how="inner")
final_dataset = nfa.merge(pop_trends, on="country", how="inner")
final_dataset
```

```python
# hdi vs pop
plt.scatter(final_dataset.iloc[:,3],final_dataset.iloc[:,7])
plt.xlabel('hdi')
plt.ylabel('population')
plt.show()


hdi = final_dataset.iloc[:,3]
population = final_dataset.iloc[:,7]
correlation = st.pearsonr(hdi,population)
print(correlation)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
# carbon produced vs pop
plt.scatter(final_dataset.iloc[:,7],final_dataset.iloc[:,4])
plt.xlabel('population')
plt.ylabel('carbon produced')
plt.show()

population = final_dataset.iloc[:,7]
carbon_produced = final_dataset.iloc[:,4]
correlation = st.pearsonr(population,carbon_produced)
print(correlation)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
final_dataset.sort_values("carbon_produced", ascending=False)
```

```python
# carbon consumed vs pop
plt.scatter(final_dataset.iloc[:,7],final_dataset.iloc[:,5])
plt.xlabel('population')
plt.ylabel('carbon consumed')
plt.show()

population = final_dataset.iloc[:,7]
carbon_consumed = final_dataset.iloc[:,7]
correlation = st.pearsonr(population,carbon_consumed)
print(correlation)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
# hdi vs hdi rank
plt.scatter(final_dataset.iloc[:,3],final_dataset.iloc[:,6])
plt.xlabel('hdi')
plt.ylabel('hdi rank')
plt.show()

hdi = final_dataset.iloc[:,3]
hdi_rank = final_dataset.iloc[:,6]
correlation = st.pearsonr(hdi,hdi_rank)
print(correlation)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```
