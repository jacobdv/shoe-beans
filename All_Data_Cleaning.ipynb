```python
# Imports and Dependencies
import pandas as pd
import numpy as np
import math as m
pd.set_option('display.max_rows', None)
nfa_data = pd.read_csv("NFA_Data.csv")
# Reads in csv for table two of the UN data, using the fifth row as header.
hdi_data = pd.read_csv("UN_Data_Table_2.csv", header=4).drop([0,67,121,159,193])
```

# NFA Data

```python
# Creating country, year, and carbon emissions dataframe for the selected years.
nfa_data = nfa_data[['country','year','carbon','record']]
nfa_data1990 = nfa_data[nfa_data['year'] == 1990]
nfa_data2000 = nfa_data[nfa_data['year'] == 2000]
nfa_data2010 = nfa_data[nfa_data['year'] == 2010]

# Creating combined dataset.
combined_years = ((nfa_data1990.append(nfa_data2000)).append(nfa_data2010))

# Dropping from country names.
combined_years = combined_years[combined_years['country'] != 'World']

# Reseting index values.
combined_years = combined_years.set_index("country").reset_index()
```

```python
# Calculating potentially important variables for merging with UN dataset.

# Number of countries.
countries_number = len(combined_years['country'].unique())

# Alphabetical list of countries.
nfa_countries_list = sorted(combined_years['country'].unique())

# Prints variables.
print(f"Number of Unique Country Names: {countries_number}")
print(f"\nList of Unique Country Names: \n{nfa_countries_list}")

# Countries that may need renaming or removal.
iffy = [] 
clear = []
reason = []
i = 0
commas = 0
spaces = 0
hyphens = 0
for name in nfa_countries_list:
    if "," in name:
        iffy.append(name)
        reason.append(1)
        commas += 1
    elif " " in name:
        iffy.append(name)
        reason.append(2)
        spaces += 1
    elif "-" in name:
        iffy.append(name)
        reason.append(3)
        hyphens += 1
    else:
        clear.append(name)
        
print(f"\n\nIffy Countries: {len(iffy)}\n{commas} Commas, {spaces} Spaces, {hyphens} Hyphens\n-------------------\nCountry   ---   Reason\n-----------------------\n")
for name in iffy:
    if reason[i] == 1:
        print(f'{i}. {iffy[i]}\n--- Comma\n')
    elif reason[i] == 2:
        print(f'{i}. {iffy[i]}\n--- Space\n')
    elif reason[i] == 3:
        print(f'{i}. {iffy[i]}\n--- Hyphen\n')
    i += 1

# Countries that have no obvious issues.
print(f"\nClear Countries: {len(clear)}\n---------------------\nCountry\n-------\n")
for name in clear:
    print(name)
```

# UN Data

```python
# Selects only columns with valid, single year names.
selected_columns = []
for column in hdi_data:
    if 'Unnamed' not in column:
        if '-' not in column:
            selected_columns.append(column)

# Turns .. into NaN values. Drops rows with only NaN values.
hdi_data = hdi_data.replace("..","NaN") 
hdi_data = (hdi_data[selected_columns])
hdi_data.dropna(thresh = 9, inplace = True)

# Drops bottom rows of summary data.
hdi_data.drop(hdi_data.tail(36).index,inplace=True)

# Transforms data from horizontal to vertical.
selected_years = pd.melt(hdi_data, id_vars = 'Country', value_vars = ['1990','2000','2010', '2014', '2015'])

# Renames columns.
selected_years = selected_years.rename(columns = {'Country':'country','variable':'year','value':'hdi'})
```

```python
# Calculating potentially important variables for merging with UN dataset.

# Number of countries.
countries_number = len(selected_years['country'].unique())

# Alphabetical list of countries.
un_countries_list = sorted(selected_years['country'].unique())

# Prints variables.
print(f"Number of Unique Country Names: {countries_number}")
print(f"\nList of Unique Country Names: \n{un_countries_list}")

# Countries that may need renaming or removal.
iffy = [] 
clear = []
reason = []
i = 0
commas = 0
spaces = 0
hyphens = 0
for name in un_countries_list:
    if "," in name:
        iffy.append(name)
        reason.append(1)
        commas += 1
    elif " " in name:
        iffy.append(name)
        reason.append(2)
        spaces += 1
    elif "-" in name:
        iffy.append(name)
        reason.append(3)
        hyphens += 1
    else:
        clear.append(name)
        
print(f"Iffy Countries: {len(iffy)}\n{commas} Commas, {spaces} Spaces, {hyphens} Hyphens\n-------------------\nCountry   ---   Reason\n-----------------------\n")
for name in iffy:
    if reason[i] == 1:
        print(f'{i}. {iffy[i]}\n--- Comma\n')
    elif reason[i] == 2:
        print(f'{i}. {iffy[i]}\n--- Space\n')
    elif reason[i] == 3:
        print(f'{i}. {iffy[i]}\n--- Hyphen\n')
    i += 1
```

```python
country_list = []
year_list = []
hdi_list = []
carbon_list = []
i = 0

for row in combined_years['country']:
    country = combined_years['country'][i]
    year = combined_years['year'][i]
    carbon = combined_years['carbon'][i]
    if country not in nfa_un_onlys:
        if country not in country_list:
            country_list.append(country)
            year_list.append(year)
            carbon_list.append(carbon) 
    i += 1
```

```python
# Retypes columns for joining.
selected_years['country'] = selected_years['country'].astype('string')
combined_years['country'] = combined_years['country'].astype('string')
selected_years['year'] = selected_years['year'].astype('int64')
combined_years['year'] = combined_years['year'].astype('int64')

# Merging datasets.
final_dataset = combined_years.merge(selected_years, on = ['country','year'])
```

```python
final_dataset.rename(columns={'carbon':'carbon_produced','record':'carbon_consumed'})
final_dataset['carbon_produced'] = combined_years['carbon'].loc[combined_years['record'] == 'EFProdPerCap']
final_dataset['carbon_consumed'] = combined_years['carbon'].loc[combined_years['record'] == 'EFConsPerCap']
final_dataset.drop(columns=['carbon'], inplace=True)
```

```python
final_dataset = final_dataset[(final_dataset['record'] == 'EFProdPerCap') | (final_dataset['record'] == 'EFConsPerCap')]
final_dataset.set_index('country').reset_index()
```

```python
to_drop = []
i = 0

for index, row in final_dataset.iterrows():
    cp = (row['carbon_produced'])
    cc = row['carbon_consumed']
    if m.isnan(cp) & m.isnan(cc):
        to_drop.append(i)
    i += 1
        
final_dataset.drop(final_dataset.index[to_drop], inplace=True)
final_dataset
```

```python
final_dataset = pd.concat([final_dataset.apply(lambda x: sorted(x, key=pd.isnull)) for _, final_dataset in final_dataset.groupby('country')]).dropna()
final_dataset.set_index('country').reset_index()
final_dataset.drop_duplicates(subset=['country','year'],keep='first',inplace=True)
final_dataset.set_index('country').reset_index()
```
